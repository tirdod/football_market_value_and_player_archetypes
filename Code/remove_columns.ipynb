```python
# Function to add age_squared and reorder columns
def add_age_squared(df):
    df['age_squared'] = df['age'] ** 2  # Calculate age_squared
    cols = list(df.columns)  # Get current columns
    age_index = cols.index('age') + 1  # Find the index after 'age'
    cols.remove('age_squared')  # Temporarily remove 'age_squared'
    cols.insert(age_index, 'age_squared')  # Insert it after 'age'
    return df[cols]  # Return DataFrame with reordered columns

# Apply the function to per_90s and per_90s
totals = add_age_squared(totals)
per_90s = add_age_squared(per_90s)
```

```python
totals.head()
```

```python
# Calculate the 'blocks' column
totals['blocks'] = totals['blocks_blocks'] + totals['blocks_sh'] + totals['blocks_pass']

# Reorder columns to place 'blocks' behind 'blocks_pass'
columns = list(totals.columns)  # Get all column names as a list
insert_position = columns.index('blocks_pass') + 1  # Find position after 'blocks_pass'
columns.remove('blocks')  # Temporarily remove 'blocks'
columns.insert(insert_position, 'blocks')  # Insert 'blocks' after 'blocks_pass'

# Reassign the DataFrame with the new column order
totals = totals[columns]

# Drop the original columns used in the calculation
totals = totals.drop(columns=['blocks_blocks', 'blocks_sh', 'blocks_pass'])

# Specify columns to drop
cols_to_drop = ['playing_time_mn/mp', 'starts_starts','starts_compl','subs_subs','team_success_ppm','team_success_ong','team_success_onga','team_success_ong','team_success_onga',
                'standard_sh','standard_sot','standard_g/sot','standard_dist','standard_pkatt','standard_fk','expected_g_xg','expected_npxg','expected_npxg/sh','expected_np:g_xg','performance_2crdy',
                'performance_off'	,'performance_crs'	,'performance_int'	,'performance_tklw'	,'performance_pkwon','performance_pkcon','performance_og',
                'performance_recov','aerial_duels_lost','tackles_tkl','tackles_def_3rd','tackles_mid_3rd','tackles_att_3rd','challenges_att','challenges_lost',
                'tklint','total_cmp','total_att','total_cmp_perc','short_cmp','short_att','short_cmp_perc','medium_cmp','medium_att','medium_cmp_perc',	
                'long_cmp','long_att','long_cmp_perc','ast_minus_xA']

# Drop the columns
totals = totals.drop(columns=cols_to_drop)

totals.head()
```
